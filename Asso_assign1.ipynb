```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
import warnings 
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv("C:\\Users\\sneha\\Downloads\\book.csv")
```

```python
df
```

```python
from mlxtend.frequent_patterns import apriori, association_rules
```

```python
score = apriori(df, min_support=0.1, use_colnames= True)
score
```

```python
model = association_rules(score, metric='lift')
model
```

```python
mod = model.iloc[:10,:]
```

```python
plt.scatter(model['support'], model['confidence'])
plt.xlabel('support')
plt.ylabel('confidence');
```

```python
plt.scatter(model['lift'], model['confidence'])
plt.xlim()
plt.xlabel('lift')
plt.ylabel('confidence');
```
